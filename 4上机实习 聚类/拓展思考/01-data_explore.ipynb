```python
import pandas as pd
```

```python
inputfile = 'air_data.csv'
outputfile = './tmp/explore.csv'
```

```python
df = pd.read_csv(inputfile)
df.head()
```

```python
explore = df.describe(percentiles=[], include='all').T
explore
```

```python
explore['null'] = len(df) - explore['count']
explore['null']
```

```python
explore = explore[['null', 'max', 'min']]
explore
```

```python
explore.columns = ['空值数', '最大值', '最小值']
```

```python
explore.to_csv(outputfile)
```
