```python
import numpy as np
import pandas as pd
from sklearn.linear_model import Lasso
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("data.csv")
df.head()
```

```python
# # 对特征进行标准化
# scaler = StandardScaler()
# X_scaled = scaler.fit_transform(df.iloc[:, 0:13])
# lasso.fit(X_scaled, df['y'])
lasso = Lasso(alpha=1000)  # 调用Lasso()函数，设置λ的值为1000
lasso.fit(df.iloc[:,0:13],df['y'])
```

```python
print("相关系数", np.around(lasso.coef_, 5))
```

```python
print('相关系数非零个数为：',np.sum(lasso.coef_ != 0))
```

```python
mask = lasso.coef_ != 0  # 返回一个相关系数是否为零的布尔数组
print('相关系数是否为零：',mask)
```

```python
outputfile ='./tmp/new_reg_data.csv'  # 输出的数据文件
new_reg_data = df.iloc[:, 0:13].iloc[:, mask] 
new_reg_data['y'] = df['y']  # 添加目标变量 'y'
new_reg_data.to_csv(outputfile, index=False)  # 存储数据
print('输出数据的维度为：',new_reg_data.shape)  # 查看输出数据的维度
```
