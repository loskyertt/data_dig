```python
from chefboost import Chefboost as chef
import pandas as pd
```

```python
df = pd.read_csv("environment_data.csv")
df.head()
```

```python
config = {'algorithm': 'C4.5'}
model = chef.fit(df, config = config, target_label = '空气等级')
```

```python
df.iloc[1]
```

```python
prediction = chef.predict(model, df.iloc[1])
prediction
```

```python
for index, instance in df.iterrows():
    prediction = chef.predict(model, instance)
    actual = instance['空气等级']

    if actual == prediction:
        classified = True
    else:
        classified = False
        print("*", end="")

    print(actual , "-", prediction)
```
